```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('economy_global.csv')
```

```python
print(data)
```

```python
print(data.info())
```

```python
d=data.describe()
d
```

```python
#min
min = data[' Population '].min()
print(f'Minimum: {min}')

#max
max = data[' Population '].max()
print(f'Maximum: {max}')

#mean
mean = data[' Population '].mean()
print(f'Mean: {mean}')

#median
median = data[' Population '].median()
print(f'Median: {median}')

#mode
mode= data[' Population '].mode()
print(f'Mode: {mode[0]}')

#standard deviation
std_dev = data[' Population '].std()
print(f'Standard Deviation: {std_dev}')
```

```python
#min
min = data[' Exports of goods and services '].min()
print(f'Minimum: {min}')

#max
max = data[' Exports of goods and services '].max()
print(f'Maximum: {max}')

#mean
mean = data[' Exports of goods and services '].mean()
print(f'Mean: {mean}')

#median
median = data[' Exports of goods and services '].median()
print(f'Median: {median}')

#mode
mode= data[' Exports of goods and services '].mode()
print(f'Mode: {mode[0]}')

#standard deviation
std_dev = data[' Exports of goods and services '].std()
print(f'Standard Deviation: {std_dev}')
```

```python
#min
min = data[' Imports of goods and services '].min()
print(f'Minimum: {min}')

#max
max = data[' Imports of goods and services '].max()
print(f'Maximum: {max}')

#mean
mean = data[' Imports of goods and services '].mean()
print(f'Mean: {mean}')

#median
median = data[' Imports of goods and services '].median()
print(f'Median: {median}')

#mode
mode= data[' Imports of goods and services '].mode()
print(f'Mode: {mode[0]}')

#standard deviation
std_dev = data[' Imports of goods and services '].std()
print(f'Standard Deviation: {std_dev}')
```

```python
#min
min = data[' General government final consumption expenditure '].min()
print(f'Minimum: {min}')

#max
max = data[' General government final consumption expenditure '].max()
print(f'Maximum: {max}')

#mean
mean = data[' General government final consumption expenditure '].mean()
print(f'Mean: {mean}')

#median
median = data[' General government final consumption expenditure '].median()
print(f'Median: {median}')

#mode
mode= data[' General government final consumption expenditure '].mode()
print(f'Mode: {mode[0]}')

#standard deviation
std_dev = data[' General government final consumption expenditure '].std()
print(f'Standard Deviation: {std_dev}')
```

```python

x = data[' Population ']
y = data[' Transport, storage and communication (ISIC I) ']

plt.scatter(x, y)
plt.xlabel('Population')
plt.ylabel('Transport, storage and communication (ISIC I)')
plt.title('Scatter Plot of Population vs. Transport, storage and communication (ISIC I)')
plt.show()
```

```python
# Calculating the IQR
quartiles = data[' Population '].quantile([0.25, 0.75])
iqr = quartiles[0.75] - quartiles[0.25]
print(iqr)
```
